```python
def matriz(hola):
    auto = np.zeros([4,4])
    i = 0
    for j in np.arange(1,4+1,1):
        auto[i,:] = np.linspace((j * hola),(j * hola) +12,4)
        i = i+1
   
    return auto
```

```python
resul = matriz(100)
print('\033[1;36m Automatic Matrix Array : \033[0m\n', resul)
pepito = matriz(1)
print('\033[1;36m Automatic Matrix Array : \033[0m\n', pepito)
```

```python
h = []
for hola in np.arange(1, 3, 1):
    h.append(matriz(hola))

z = np.vstack(h)
```

```python
print('\033[1;36m Automatic Matrix Array : \033[0m\n', z)
```

#### Stiffness Matrix in Local Coordinates (RMF)
![alt text](image.png)
![alt text](image-1.png)

```python
A = 0.3*0.3
I = 0.3**4/12
E = 2000000
G = 15000
L = 5
f = 6/5
```

```python
def stiffness_matrix_Element_RMF_L(A,I,E,L,G,f):
    Be = 6*E*I*f / (G*A*L**2)
    r = A*E /L
    kp = 2*E*I/L * (2 + Be)/(1 + 2*Be)
    ap = 2*E*I/L * (1 - Be)/(1 - 2*Be)
    bp = 6*E*I/(L**2) * (1)/(1 + 2*Be)
    tp = 12*E*I/(L**3) * (1)/(1 + 2*Be)

    kl = np.array([[r,0,0,-r,0,0],
                [0,tp,bp,0,-tp,bp],
                [0,bp,kp,0,-bp,ap],
                [-r,0,0,r,0,0],
                [0,-tp,-bp,0,tp,-bp],
                [0,bp,ap,0,-bp,kp]], dtype=float)
    
    return kl
```

```python
pepito = stiffness_matrix_Element_RMF_L(2*A,I,E,L/2,G,f)
pepito_df = pd.DataFrame(pepito)
pepito_df
```
